# Training

```python
import os

repo_path = "/home/michel/Documents/GitHub/CardDetection"
label = "labelmap.pbtxt"
record = "train.record"
num_classes = 9

config_path = os.path.join(repo_path, "samples/training/cfg")

file = open(os.path.join(config_path, "pipeline.config"),"r",encoding="UTF8")
content = file.read()
file.close()

## Set the number of classes
MODEL_FULL_PATH = os.path.join(repo_path, "src/models/faster_rcnn_inception_v2_coco_2018_01_28/model.ckpt")
LABEL_FULL_PATH = os.path.join(config_path, label)
RECORD_FULL_PATH  = os.path.join(config_path, record)
NUM_CLASSES_STR  = str(num_classes)

content = content.replace("NUM_CLASSES", NUM_CLASSES_STR)
content = content.replace("MODEL_FULL_PATH","\"{}\"".format(MODEL_FULL_PATH))
content = content.replace("LABEL_FULL_PATH","\"{}\"".format(LABEL_FULL_PATH))
content = content.replace("RECORD_FULL_PATH","\"{}\"".format(RECORD_FULL_PATH))

file = open(os.path.join(config_path, "train.config"),"w+",encoding="UTF8")
file.write(content)
file.close()
```
